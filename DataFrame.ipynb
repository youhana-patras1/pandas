Creating Dataframe Python Dictionary

```python
data_dict = {
    'Name': ['Ali', 'Ahmed', 'Sara', 'Zain', 'Ayesha'],
    'City': ['Karachi', 'Lahore', 'Islamabad', 'Multan', 'Peshawar'],
    'Salary': [55000, 62000, 75000, 48000, 51000]
}
df = pd.DataFrame(data_dict)
df
```

Creating Dataframe Python List

```python
data_list = [
    ['Ali', 'Karachi', 55000],
    ['Ahmed', 'Lahore', 62000],
    ['Sara', 'Islamabad', 75000],
    ['Zain', 'Multan', 48000],
    ['Ayesha', 'Peshawar', 51000]
]
df = pd.DataFrame(data_list)
df
```

```python
data_list = [
    ['Ali', 'Karachi', 55000],
    ['Ahmed', 'Lahore', 62000],
    ['Sara', 'Islamabad', 75000],
    ['Zain', 'Multan', 48000],
    ['Ayesha', 'Peshawar', 51000]
]
df = pd.DataFrame(data_list)
# Columns ke naam sath de dein
df = pd.DataFrame(data_list,columns=['Name','City','Salary'])
df
```

```python
# Jab aap df['Name'] likhte hain
# to Pandas poore table mein se sirf 'Name' wala column nikaal kar aapko dikhata hai.
df['Name']
```

Multiple Columns Select

```python
# "Multiple Column Selection" kehlati hai. Jab aap double square brackets [[...]] 
# istemal karte hain, to Pandas aik naya DataFrame (chota table) nikaal kar deta hai
#jis mein sirf aap ke pasandida columns hote hain.
df[['Name','City']]
```

Conditional Selection

```python
df
```

'Salary' wale column ki har value ko check karta hai aur ek True/False ki list (Series) banata hai.
Agar Salary 62,000 hai -> True
Agar Salary 55,000 hai -> False

```python
df[df["Salary"] > 60000]
```

Do Conditions Ek Saath Agar aapko wo log chahiye jin ki Salary > 60,000 ho aur wo 'Islamabad' mein rehte hon

```python
df[(df["Salary"] > 60000) & (df["City"] == 'Islamabad')]
```
