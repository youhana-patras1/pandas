Finding Missing Data

```python
# Dictionary taiyar karein (5 records ke saath)
data_dict = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Ali', 'Ahmed', 'Sara', 'Zain', 'Ayesha'],
    'City': ['Karachi', 'Lahore', 'Islamabad', 'Multan', 'Peshawar'],
    'Salary': [55000, 62000, np.nan, 48000, 51000], # Ek value missing (NaN) rakhi hai
    'Designation': ['Junior Dev', 'Backend Dev', 'Senior Dev', 'DevOps', 'Architect']
}
# Data ko table bana do
df = pd.DataFrame(data_dict)
# Table ko check karein
df
```

```python
# Har column mein kitni values missing hain
df.isnull().sum()
```

Removing Missing Data

```python
# Yeh function wo rows delete kar deta hai jahan koi bhi value missing (NaN) ho.
df.dropna()
```
